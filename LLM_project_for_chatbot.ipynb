```python
import os
import torch
import pandas as pd
try:
    import bitsandbytes as bnb
except ImportError:
    print("bitsandbytes not found. Please install using: pip install bitsandbytes")
    bnb = None

from datasets import Dataset
from transformers import (
    AutoModelForCausalLM,
    AutoTokenizer,
    BitsAndBytesConfig,
    TrainingArguments,
)
from peft import LoraConfig
from trl import SFTTrainer, SFTConfig

# Model Configuration
model_name = "unsloth/Llama-3.2-1B-Instruct"
new_model = "kerala-agriculture-multilingual-quick"

# Reduced Training Parameters
num_train_epochs = 5  # Reduced from 30
per_device_train_batch_size = 1  # Smaller batch size
learning_rate = 3e-4  # Slight adjustment
max_seq_length = 256  # Reduced sequence length

# Output Configuration
output_dir = "./quick_multilingual_results"

# Device Configuration
device_map = {"": 0}  # Specify first GPU

def prepare_multilingual_dataset(file_path, sample_size=100):
    """
    Prepare dataset with multilingual prompts
    Supports both Malayalam and English inputs
    """
    # Read CSV file and sample
    df = pd.read_csv(file_path).head(sample_size)

    # Create multilingual prompts
    def create_multilingual_prompt(row):
        input_text = (
            f"Crop: {row.get('crop_name_ml', 'N/A')} ({row.get('crop_name_en', 'N/A')}) "
            f"Zones: {row.get('cultivation_zones', 'N/A')} "
            f"Soil Types: {row.get('soil_types', 'N/A')} "
            f"Fertilizers: {row.get('fertilizers', 'N/A')} "
        )

        output_text = (
            f"Cultivation details for {row.get('crop_name_ml', 'crop')}: "
            f"Best in {row.get('cultivation_zones', 'various')} regions. "
            f"Suitable soils: {row.get('soil_types', 'multiple')}. "
            f"Recommended fertilizers: {row.get('fertilizers', 'standard')}."
        )

        return f"### Input:\n{input_text}\n\n### Response:\n{output_text}"

    # Generate prompts
    df['formatted_prompt'] = df.apply(create_multilingual_prompt, axis=1)

    return df['formatted_prompt'].tolist()

def fine_tune_quick_model(dataset_path):
    """
    Quick fine-tuning of Llama multilingual agricultural model
    """
    try:
        # Quantization Configuration (with fallback)
        if bnb and torch.cuda.is_available():
            bnb_config = BitsAndBytesConfig(
                load_in_4bit=True,
                bnb_4bit_quant_type="nf4",
                bnb_4bit_compute_dtype=torch.float16,
                bnb_4bit_use_double_quant=True
            )
            load_in_4bit = True
        else:
            bnb_config = None
            load_in_4bit = False
            print("Using standard loading due to CPU environment or missing bitsandbytes")

        # Load base model
        model = AutoModelForCausalLM.from_pretrained(
            model_name,
            quantization_config=bnb_config if load_in_4bit else None,
            device_map=device_map if torch.cuda.is_available() else "cpu"
        )
        model.config.use_cache = False

        # Load tokenizer
        tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
        tokenizer.pad_token = tokenizer.eos_token
        tokenizer.padding_side = "right"

        # Prepare dataset
        dataset_texts = prepare_multilingual_dataset(dataset_path)
        dataset = Dataset.from_dict({"text": dataset_texts})

        # Simplified LoRA Configuration
        peft_config = LoraConfig(
            r=16,  # Reduced rank
            lora_alpha=32,
            lora_dropout=0.1,
            bias="none",
            task_type="CAUSAL_LM",
            target_modules=[
                "q_proj", "k_proj", "v_proj", "o_proj",
                "gate_proj", "up_proj", "down_proj"
            ]
        )

        # Training Arguments (Streamlined)
        training_arguments = TrainingArguments(
            output_dir=output_dir,
            num_train_epochs=num_train_epochs,
            per_device_train_batch_size=per_device_train_batch_size,
            learning_rate=learning_rate,
            weight_decay=0.001,
            logging_steps=10,
            save_strategy="epoch",
            fp16=torch.cuda.is_available(),
            bf16=False,
            report_to="none"  # Disable external logging
        )

        # Trainer Configuration
        trainer = SFTTrainer(
            model=model,
            train_dataset=dataset,
            peft_config=peft_config,
            dataset_text_field="text",
            max_seq_length=max_seq_length,
            tokenizer=tokenizer,
            args=training_arguments,
        )

        # Start quick training
        trainer.train()

        # Save model
        trainer.model.save_pretrained(new_model)
        tokenizer.save_pretrained(new_model)

        return model, tokenizer

    except Exception as e:
        print(f"An error occurred during model training: {e}")
        return None, None

# Main Execution
if __name__ == "__main__":
    # Path to your agricultural dataset
    dataset_path = "/content/kerala_agricultural_dataset.csv"

    # Perform quick fine-tuning
    model, tokenizer = fine_tune_quick_model(dataset_path)

    if model and tokenizer:
        print("Quick Multilingual Agriculture Model Fine-Tuning Completed!")
    else:
        print("Model fine-tuning failed.")
```
